```python
df.drop('phishing', axis=1, inplace=True)
dct = {}
dct_2 = {}
for h in df.columns:
    dct[str(h.split('_')[0]+' of '+h.split('_')[1]).replace('qty','Quantity')] = sum(df[h])
    if h!='length_url':
        dct_2[str(h.split('_')[0]+' of '+h.split('_')[1]).replace('qty','Quantity')] = sum(df[h])
```

```python
import itertools
plt.figure(figsize=(8,5))
temp = sorted(dct.items(), key=lambda x: x[1],reverse=True)
new_dct = {}
c = 0
for g in temp:
    title,value = g
    if c<=4:
        new_dct[title] = value
        c+=1

new_dct = dict(sorted(new_dct.items(), key=lambda item: item[1],reverse = True))
new_dct = dict(itertools.islice(new_dct.items(), 7))

ax = sns.barplot(x = list(new_dct.values()),
            y = list(new_dct.keys()))
ax.set_title('The features of most repetitive phishing websites')
new_dct = dict(sorted(new_dct.items(), key=lambda item: item[1],reverse = True))
plt.show()
```
